# Readout Emptying Qualification Workflow

This notebook is the advanced-protocol companion to the segmented readout-emptying seed API in `cqed_sim.optimal_control`.

It is intentionally workflow-oriented:

1. build the analytic seed,
2. compare it to a square pulse,
3. replay it under Kerr and Lindblad models,
4. refine it with the reduced outer-loop harness,
5. and inspect the checked-in study artifacts generated by the scripts under `examples/studies/readout_emptying/`.


## Source-of-truth scripts

- `examples/studies/readout_emptying/linear_seed_validation.py`
- `examples/studies/readout_emptying/kerr_replay_and_chirp.py`
- `examples/studies/readout_emptying/dispersive_lindblad_validation.py`
- `examples/studies/readout_emptying/reduced_refinement.py`
- `examples/studies/readout_emptying/hardware_sensitivity.py`
- `examples/studies/readout_emptying/summary_benchmark.py`

These scripts write their evidence under `outputs/readout_emptying_qualification/` and refresh the tutorial-ready figures in `documentations/assets/images/tutorials/readout_emptying/`.


```python
from pathlib import Path
import json

summary_path = Path("outputs/readout_emptying_qualification/05_summary_benchmark/metrics.json")
if summary_path.exists():
    payload = json.loads(summary_path.read_text(encoding="utf-8"))
    for label, metrics in payload["comparison_table"].items():
        print(
            label,
            f"residual={metrics['max_final_residual_photons']:.3e}",
            f"accuracy={metrics.get('measurement_chain_accuracy', float('nan')):.3f}",
            f"non_qnd={metrics.get('non_qnd_total', float('nan')):.3e}",
        )
else:
    print("Run examples/studies/readout_emptying/summary_benchmark.py to populate the checked-in summary metrics.")
```
